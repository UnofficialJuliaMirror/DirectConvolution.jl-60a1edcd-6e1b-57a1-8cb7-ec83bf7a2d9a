```julia
using Revise
using DirectConvolution
using Plots
```

```julia
signal = sin.(1:0.3:10)
filter = UDWT_Filter_Haar{Float64}()
c = udwt(signal,filter,scale=1)
reconst = Array{Float64,1}(length(signal))
inverse_udwt!(c,reconst)
plot(signal,label="signal")
plot!(c.V,label="V")
```

```julia
plot(signal,label="signal")
plot!(reconst,label="recons")
```
